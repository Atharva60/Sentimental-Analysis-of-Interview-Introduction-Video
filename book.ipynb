```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_excel('compare.xlsx')
df
```

```python
columns_to_plot = df.columns[1:]  # Exclude the 'Student' column
for column in columns_to_plot:
    plt.figure(figsize=(10, 5))
    plt.plot(df['Student'], df[column], marker='o', linestyle='-')
    plt.xlabel('Student')
    plt.ylabel(column)
    plt.title(f'{column} vs. Student')
    plt.grid(True)
    plt.show()
```
